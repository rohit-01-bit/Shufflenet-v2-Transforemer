```python
import os
import numpy as np
import tensorflow as tf
import cv2
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Dropout, GlobalAveragePooling2D, Concatenate, Lambda, GlobalAveragePooling1D, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, roc_curve, confusion_matrix, classification_report
from imblearn.over_sampling import RandomOverSampler
import gc
from tensorflow.keras.losses import BinaryCrossentropy

# Constants
IMAGE_SIZE = (224, 224)
BATCH_SIZE = 32  # Reduced batch size to save memory
EPOCHS = 50
LEARNING_RATE = 1e-4
SMOOTHING = 0.2  # Label smoothing factor

# Enable mixed precision training
tf.keras.mixed_precision.set_global_policy('mixed_float16')

# Restrict TensorFlow to only use needed GPU memory
gpus = tf.config.list_physical_devices('GPU')
if gpus:
    try:
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
    except RuntimeError as e:
        print(e)

# Clear memory
gc.collect()
tf.keras.backend.clear_session()

# Paths
authentic_paths = [
    r"C:\Users\Rohit Vardhan\Documents\EOC & MFC\datasets\Jphash_modified_CASIA_upload[extracted]\training"
]
tampered_paths = [
    r"C:\Users\Rohit Vardhan\Documents\EOC & MFC\datasets\Jphash_modified_CASIA_upload[extracted]\testing"
]

# Load Data
def collect_image_files(authentic_paths, tampered_paths):
    authentic_files, tampered_files = [], []
    for path in authentic_paths + tampered_paths:
        for subdir, _, files in os.walk(path):
            for file in files:
                if file.lower().endswith(('.png', '.jpeg', '.jpg', '.tif')):
                    full_path = os.path.join(subdir, file)
                    (authentic_files if path in authentic_paths else tampered_files).append(full_path)
    print(f"Number of authentic images found: {len(authentic_files)}")
    print(f"Number of tampered images found: {len(tampered_files)}")
    return authentic_files, tampered_files

authentic_files, tampered_files = collect_image_files(authentic_paths, tampered_paths)
all_files = authentic_files + tampered_files
all_labels = [0] * len(authentic_files) + [1] * len(tampered_files)

print(f"Total images before oversampling: {len(all_files)}")

# Handle Imbalance
oversampler = RandomOverSampler(random_state=42)
all_files, all_labels = oversampler.fit_resample(np.array(all_files).reshape(-1, 1), np.array(all_labels))
all_files = all_files.flatten()

print(f"Total images after oversampling: {len(all_files)}")

# Split data
train_files, test_files, train_labels, test_labels = train_test_split(
    all_files, all_labels, test_size=0.2, random_state=42, stratify=all_labels)

print(f"Training images: {len(train_files)}")
print(f"Testing images: {len(test_files)}")

# Data Generator with Augmentation
class CustomImageDataGenerator(tf.keras.utils.Sequence):
    def __init__(self, image_files, labels, batch_size, image_size, shuffle=True, augment=False):
        self.image_files = np.array(image_files)
        self.labels = np.array(labels).reshape(-1, 1)
        self.batch_size = batch_size
        self.image_size = image_size
        self.shuffle = shuffle
        self.augment = augment
        self.on_epoch_end()

    def __len__(self):
        return int(np.ceil(len(self.image_files) / self.batch_size))

    def __getitem__(self, idx):
        batch_indices = self.indices[idx * self.batch_size:(idx + 1) * self.batch_size]
        batch_images, batch_labels = [], []
        for i in batch_indices:
            img = cv2.imread(self.image_files[i], cv2.IMREAD_COLOR)  # Read as color image
            if img is not None:
                img = cv2.resize(img, self.image_size)
                if self.augment:
                    img = self.augment_image(img)  # Apply augmentation
                batch_images.append(img)
                batch_labels.append(self.labels[i])
        return np.array(batch_images, dtype=np.float16) / 255.0, np.array(batch_labels)

    def augment_image(self, img):
        img = tf.image.random_flip_left_right(img)
        img = tf.image.random_brightness(img, max_delta=0.1)
        img = tf.image.random_contrast(img, lower=0.8, upper=1.2)
        return img

    def on_epoch_end(self):
        self.indices = np.arange(len(self.image_files))
        if self.shuffle:
            np.random.shuffle(self.indices)

train_gen = CustomImageDataGenerator(train_files, train_labels, BATCH_SIZE, IMAGE_SIZE, augment=True)
test_gen = CustomImageDataGenerator(test_files, test_labels, BATCH_SIZE, IMAGE_SIZE, shuffle=False)

def extract_from_generator(generator):
    X, y = [], []
    
    for batch_x, batch_y in generator:
        X.extend(batch_x)
        y.extend(batch_y)

        # Break if we've gathered enough samples
        if len(X) >= len(generator.labels):  # Use labels length as a stopping condition
            break
    
    return np.array(X), np.array(y)
    
X_train, y_train = extract_from_generator(train_gen)
X_test, y_test = extract_from_generator(test_gen)




import numpy as np
import keras.backend as K
from keras.layers import Input, Conv2D, MaxPool2D, DepthwiseConv2D, BatchNormalization, Activation, Concatenate, GlobalAveragePooling2D, Dense, Lambda, Flatten, Reshape, MultiHeadAttention, Dropout, LayerNormalization, Add
from keras.models import Model

# Function for Channel Shuffle
def channel_shuffle(x):
    height, width, channels = x.shape.as_list()[1:]
    channels_per_split = channels // 2
    x = tf.reshape(x, [-1, height, width, 2, channels_per_split])
    x = tf.transpose(x, perm=(0, 1, 2, 4, 3))
    x = tf.reshape(x, [-1, height, width, channels])
    return x

# ShuffleNet Unit (Shuffle Block)
def shuffle_unit(inputs, out_channels, bottleneck_ratio, strides=2, stage=1, block=1):
    if K.image_data_format() == 'channels_last':
        bn_axis = -1
    else:
        raise ValueError('Only channels last supported')

    prefix = f'stage{stage}_block{block}'
    bottleneck_channels = int(out_channels * bottleneck_ratio)
    
    x = Conv2D(bottleneck_channels, kernel_size=(1,1), strides=1, padding='same', name=f'{prefix}_1x1conv_1')(inputs)
    x = BatchNormalization(axis=bn_axis, name=f'{prefix}_bn_1x1conv_1')(x)
    x = Activation('relu', name=f'{prefix}_relu_1x1conv_1')(x)
    
    x = DepthwiseConv2D(kernel_size=3, strides=strides, padding='same', name=f'{prefix}_3x3dwconv')(x)
    x = BatchNormalization(axis=bn_axis, name=f'{prefix}_bn_3x3dwconv')(x)
    
    x = Conv2D(bottleneck_channels, kernel_size=1, strides=1, padding='same', name=f'{prefix}_1x1conv_2')(x)
    x = BatchNormalization(axis=bn_axis, name=f'{prefix}_bn_1x1conv_2')(x)
    x = Activation('relu', name=f'{prefix}_relu_1x1conv_2')(x)

    ret = Lambda(channel_shuffle, output_shape=lambda s: s, name=f'{prefix}_channel_shuffle')(x)
    return ret

# Transformer Block
def TransformerBlock(inputs, num_heads=8, ff_dim=512, dropout_rate=0.1):
    attn_output = MultiHeadAttention(num_heads=num_heads, key_dim=inputs.shape[-1] // num_heads)(inputs, inputs)
    attn_output = Dropout(dropout_rate)(attn_output)
    out1 = Add()([inputs, attn_output])
    out1 = LayerNormalization(epsilon=1e-6)(out1)
    
    ffn_output = Dense(ff_dim, activation='relu')(out1)
    ffn_output = Dense(inputs.shape[-1])(ffn_output)  # Match input shape
    ffn_output = Dropout(dropout_rate)(ffn_output)
    
    out2 = Add()([out1, ffn_output])
    out2 = LayerNormalization(epsilon=1e-6)(out2)
    return out2

# ShuffleNetV2 Backbone
def ShuffleNetV2(input_shape=(32,32,3)):
    img_input = Input(shape=input_shape)
    
    x = Conv2D(24, kernel_size=(3, 3), padding='same', use_bias=False, strides=(2, 2), activation='relu', name='conv1')(img_input)
    x = MaxPool2D(pool_size=(3, 3), strides=(2, 2), padding='same', name='maxpool1')(x)
    x = shuffle_unit(x, out_channels=116, strides=2, bottleneck_ratio=1, stage=2, block=1)
    
    x = Conv2D(1024, kernel_size=1, padding='same', strides=1, activation='relu', name='conv_final')(x)
    x = GlobalAveragePooling2D(name='global_avg_pool')(x)
    
    return Model(img_input, x, name='ShuffleNetV2')

# Hybrid Model with ShuffleNet and Transformer
def HybridModel(input_shape=(224, 224, 3)):  # Changed to 3 channels
    shufflenet = ShuffleNetV2(input_shape=input_shape)
    shufflenet_output = shufflenet.output  # (batch_size, features)

    # Reshape for Transformer block
    reshaped_output = Reshape((1, shufflenet_output.shape[-1]))(shufflenet_output)

    # Pass through Transformer block
    transformer_output = TransformerBlock(reshaped_output)

    # Flatten and classification layers
    transformer_output = Flatten()(transformer_output)
    x = Dense(512, activation='relu')(transformer_output)
    x = Dropout(0.3)(x)
    x = Dense(1, activation='sigmoid')(x)  # Single output neuron with sigmoid activation

    model = Model(inputs=shufflenet.input, outputs=x)
    return model

# Create and compile the model
model = HybridModel(input_shape=(224, 224, 3))  # Changed to 3 channels
model.compile(
    optimizer=Adam(learning_rate=LEARNING_RATE),
    loss=BinaryCrossentropy(label_smoothing=SMOOTHING),  # Use built-in function
    metrics=['accuracy', tf.keras.metrics.AUC(name='auc')]
)

# Print model summary
print(model.summary())

# Callbacks
class ClearMemory(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs=None):
        gc.collect()
        tf.keras.backend.clear_session()

clear_memory = ClearMemory()

early_stopping = EarlyStopping(monitor='val_auc', patience=10, restore_best_weights=True, mode='max')
reduce_lr = ReduceLROnPlateau(monitor='val_auc', factor=0.5, patience=5, min_lr=1e-7, verbose=1, mode='max')
model_checkpoint = ModelCheckpoint('best_hybrid_model.keras', monitor='val_auc', mode='max', save_best_only=True, verbose=1)

# Train Model
history = model.fit(
    train_gen,
    validation_data=test_gen,
    epochs=EPOCHS,
    callbacks=[early_stopping, reduce_lr, model_checkpoint, clear_memory]
)

# Save the model
model.save('hybrid_model.keras')
print("Model saved as hybrid_model.keras")

# Evaluate Model
test_preds = model.predict(test_gen)
test_preds = (test_preds > 0.5).astype(int)  # Convert probabilities to binary predictions

# Calculate metrics
accuracy = accuracy_score(test_labels, test_preds)
precision = precision_score(test_labels, test_preds)
recall = recall_score(test_labels, test_preds)
f1 = f1_score(test_labels, test_preds)
roc_auc = roc_auc_score(test_labels, test_preds)

# Confusion Matrix
conf_matrix = confusion_matrix(test_labels, test_preds)

# Classification Report
class_report = classification_report(test_labels, test_preds, target_names=['Authentic', 'Tampered'])

# ROC Curve
fpr, tpr, thresholds = roc_curve(test_labels, test_preds)

# AUC Curve plotting
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f'AUC Curve (AUC = {roc_auc:.4f})', color='blue')
plt.plot([0, 1], [0, 1], 'k--', label='Random Guess')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('AUC Curve')
plt.legend(loc='lower right')
plt.grid(alpha=0.3)
plt.show()

# Print all results
print("\n=== Evaluation Metrics ===")
print(f"Test Accuracy: {accuracy:.4f}")
print(f"Test Precision: {precision:.4f}")
print(f"Test Recall: {recall:.4f}")
print(f"Test F1-Score: {f1:.4f}")
print(f"Test ROC-AUC: {roc_auc:.4f}")

print("\n=== Confusion Matrix ===")
print(conf_matrix)

print("\n=== Classification Report ===")
print(class_report)

# Plot Confusion Matrix
plt.figure(figsize=(6, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['Authentic', 'Tampered'], yticklabels=['Authentic', 'Tampered'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()

# Plot Precision, Recall, and F1-Score in a bar graph
metrics = ['Precision', 'Recall', 'F1-Score']
values = [precision, recall, f1]

plt.figure(figsize=(8, 5))
bars = plt.bar(metrics, values, color=['blue', 'green', 'orange'])

# Add values on top of the bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f'{height:.4f}',
             ha='center', va='bottom', fontsize=10)

plt.title('Precision, Recall, and F1-Score')
plt.xlabel('Metrics')
plt.ylabel('Score')
plt.ylim(0, 1)  # Set y-axis limits between 0 and 1
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Plot Model Loss and Accuracy over Epochs
plt.figure(figsize=(10, 5))

# Plot Train vs Validation Accuracy
plt.figure(figsize=(8,5))
plt.plot(history.history['accuracy'], label='Train Accuracy', color='blue')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy', color='red')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Train vs Validation Accuracy')
plt.legend()
plt.grid()
plt.show()

# Plot Train vs Validation Loss
plt.figure(figsize=(8,5))
plt.plot(history.history['loss'], label='Train Loss', color='blue')
plt.plot(history.history['val_loss'], label='Validation Loss', color='red')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Train vs Validation Loss')
plt.legend()
plt.grid()
plt.show()

plt.tight_layout()
plt.show()
```

```python
import os
import numpy as np
import tensorflow as tf
import cv2
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Dropout, GlobalAveragePooling2D, Concatenate, Lambda, GlobalAveragePooling1D, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, roc_curve, confusion_matrix, classification_report
from imblearn.over_sampling import RandomOverSampler
import gc
from tensorflow.keras.losses import BinaryCrossentropy

# Constants
IMAGE_SIZE = (224, 224)
BATCH_SIZE = 32  # Reduced batch size to save memory
EPOCHS = 1
LEARNING_RATE = 1e-4
SMOOTHING = 0.2  # Label smoothing factor
SMOOTHING_FACTOR = 0.8  # Smoothing factor for plotting

# Enable mixed precision training
tf.keras.mixed_precision.set_global_policy('mixed_float16')

# Restrict TensorFlow to only use needed GPU memory
gpus = tf.config.list_physical_devices('GPU')
if gpus:
    try:
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
    except RuntimeError as e:
        print(e)

# Clear memory
gc.collect()
tf.keras.backend.clear_session()

# Paths
authentic_paths = [
    r"C:\Users\Rohit Vardhan\Documents\EOC & MFC\datasets\Jphash_modified_CASIA_upload[extracted]\training"
]
tampered_paths = [
    r"C:\Users\Rohit Vardhan\Documents\EOC & MFC\datasets\Jphash_modified_CASIA_upload[extracted]\testing"
]

# Load Data
def collect_image_files(authentic_paths, tampered_paths):
    authentic_files, tampered_files = [], []
    for path in authentic_paths + tampered_paths:
        for subdir, _, files in os.walk(path):
            for file in files:
                if file.lower().endswith(('.png', '.jpeg', '.jpg', '.tif')):
                    full_path = os.path.join(subdir, file)
                    (authentic_files if path in authentic_paths else tampered_files).append(full_path)
    print(f"Number of authentic images found: {len(authentic_files)}")
    print(f"Number of tampered images found: {len(tampered_files)}")
    return authentic_files, tampered_files

authentic_files, tampered_files = collect_image_files(authentic_paths, tampered_paths)
all_files = authentic_files + tampered_files
all_labels = [0] * len(authentic_files) + [1] * len(tampered_files)

print(f"Total images before oversampling: {len(all_files)}")

# Handle Imbalance
oversampler = RandomOverSampler(random_state=42)
all_files, all_labels = oversampler.fit_resample(np.array(all_files).reshape(-1, 1), np.array(all_labels))
all_files = all_files.flatten()

print(f"Total images after oversampling: {len(all_files)}")

# Split data
train_files, test_files, train_labels, test_labels = train_test_split(
    all_files, all_labels, test_size=0.2, random_state=42, stratify=all_labels)

print(f"Training images: {len(train_files)}")
print(f"Testing images: {len(test_files)}")

# Data Generator with Augmentation
class CustomImageDataGenerator(tf.keras.utils.Sequence):
    def __init__(self, image_files, labels, batch_size, image_size, shuffle=True, augment=False):
        self.image_files = np.array(image_files)
        self.labels = np.array(labels).reshape(-1, 1)
        self.batch_size = batch_size
        self.image_size = image_size
        self.shuffle = shuffle
        self.augment = augment
        self.on_epoch_end()

    def __len__(self):
        return int(np.ceil(len(self.image_files) / self.batch_size))

    def __getitem__(self, idx):
        batch_indices = self.indices[idx * self.batch_size:(idx + 1) * self.batch_size]
        batch_images, batch_labels = [], []
        for i in batch_indices:
            img = cv2.imread(self.image_files[i], cv2.IMREAD_COLOR)  # Read as color image
            if img is not None:
                img = cv2.resize(img, self.image_size)
                if self.augment:
                    img = self.augment_image(img)  # Apply augmentation
                batch_images.append(img)
                batch_labels.append(self.labels[i])
        return np.array(batch_images, dtype=np.float16) / 255.0, np.array(batch_labels)

    def augment_image(self, img):
        img = tf.image.random_flip_left_right(img)
        img = tf.image.random_brightness(img, max_delta=0.1)
        img = tf.image.random_contrast(img, lower=0.8, upper=1.2)
        return img

    def on_epoch_end(self):
        self.indices = np.arange(len(self.image_files))
        if self.shuffle:
            np.random.shuffle(self.indices)

train_gen = CustomImageDataGenerator(train_files, train_labels, BATCH_SIZE, IMAGE_SIZE, augment=True)
test_gen = CustomImageDataGenerator(test_files, test_labels, BATCH_SIZE, IMAGE_SIZE, shuffle=False)

def extract_from_generator(generator):
    X, y = [], []
    
    for batch_x, batch_y in generator:
        X.extend(batch_x)
        y.extend(batch_y)

        # Break if we've gathered enough samples
        if len(X) >= len(generator.labels):  # Use labels length as a stopping condition
            break
    
    return np.array(X), np.array(y)
    
X_train, y_train = extract_from_generator(train_gen)
X_test, y_test = extract_from_generator(test_gen)

import numpy as np
import keras.backend as K
from keras.layers import Input, Conv2D, MaxPool2D, DepthwiseConv2D, BatchNormalization, Activation, Concatenate, GlobalAveragePooling2D, Dense, Lambda, Flatten, Reshape, MultiHeadAttention, Dropout, LayerNormalization, Add
from keras.models import Model

# Function for Channel Shuffle
def channel_shuffle(x):
    height, width, channels = x.shape.as_list()[1:]
    channels_per_split = channels // 2
    x = tf.reshape(x, [-1, height, width, 2, channels_per_split])
    x = tf.transpose(x, perm=(0, 1, 2, 4, 3))
    x = tf.reshape(x, [-1, height, width, channels])
    return x

# ShuffleNet Unit (Shuffle Block)
def shuffle_unit(inputs, out_channels, bottleneck_ratio, strides=2, stage=1, block=1):
    if K.image_data_format() == 'channels_last':
        bn_axis = -1
    else:
        raise ValueError('Only channels last supported')

    prefix = f'stage{stage}_block{block}'
    bottleneck_channels = int(out_channels * bottleneck_ratio)
    
    x = Conv2D(bottleneck_channels, kernel_size=(1,1), strides=1, padding='same', name=f'{prefix}_1x1conv_1')(inputs)
    x = BatchNormalization(axis=bn_axis, name=f'{prefix}_bn_1x1conv_1')(x)
    x = Activation('relu', name=f'{prefix}_relu_1x1conv_1')(x)
    
    x = DepthwiseConv2D(kernel_size=3, strides=strides, padding='same', name=f'{prefix}_3x3dwconv')(x)
    x = BatchNormalization(axis=bn_axis, name=f'{prefix}_bn_3x3dwconv')(x)
    
    x = Conv2D(bottleneck_channels, kernel_size=1, strides=1, padding='same', name=f'{prefix}_1x1conv_2')(x)
    x = BatchNormalization(axis=bn_axis, name=f'{prefix}_bn_1x1conv_2')(x)
    x = Activation('relu', name=f'{prefix}_relu_1x1conv_2')(x)

    ret = Lambda(channel_shuffle, output_shape=lambda s: s, name=f'{prefix}_channel_shuffle')(x)
    return ret

# Transformer Block
def TransformerBlock(inputs, num_heads=8, ff_dim=512, dropout_rate=0.1):
    attn_output = MultiHeadAttention(num_heads=num_heads, key_dim=inputs.shape[-1] // num_heads)(inputs, inputs)
    attn_output = Dropout(dropout_rate)(attn_output)
    out1 = Add()([inputs, attn_output])
    out1 = LayerNormalization(epsilon=1e-6)(out1)
    
    ffn_output = Dense(ff_dim, activation='relu')(out1)
    ffn_output = Dense(inputs.shape[-1])(ffn_output)  # Match input shape
    ffn_output = Dropout(dropout_rate)(ffn_output)
    
    out2 = Add()([out1, ffn_output])
    out2 = LayerNormalization(epsilon=1e-6)(out2)
    return out2

# ShuffleNetV2 Backbone
def ShuffleNetV2(input_shape=(32,32,3)):
    img_input = Input(shape=input_shape)
    
    x = Conv2D(24, kernel_size=(3, 3), padding='same', use_bias=False, strides=(2, 2), activation='relu', name='conv1')(img_input)
    x = MaxPool2D(pool_size=(3, 3), strides=(2, 2), padding='same', name='maxpool1')(x)
    x = shuffle_unit(x, out_channels=116, strides=2, bottleneck_ratio=1, stage=2, block=1)
    
    x = Conv2D(1024, kernel_size=1, padding='same', strides=1, activation='relu', name='conv_final')(x)
    x = GlobalAveragePooling2D(name='global_avg_pool')(x)
    
    return Model(img_input, x, name='ShuffleNetV2')

# Hybrid Model with ShuffleNet and Transformer
def HybridModel(input_shape=(224, 224, 3)):  # Changed to 3 channels
    shufflenet = ShuffleNetV2(input_shape=input_shape)
    shufflenet_output = shufflenet.output  # (batch_size, features)

    # Reshape for Transformer block
    reshaped_output = Reshape((1, shufflenet_output.shape[-1]))(shufflenet_output)

    # Pass through Transformer block
    transformer_output = TransformerBlock(reshaped_output)

    # Flatten and classification layers
    transformer_output = Flatten()(transformer_output)
    x = Dense(512, activation='relu')(transformer_output)
    x = Dropout(0.3)(x)
    x = Dense(1, activation='sigmoid')(x)  # Single output neuron with sigmoid activation

    model = Model(inputs=shufflenet.input, outputs=x)
    return model

# Create and compile the model
model = HybridModel(input_shape=(224, 224, 3))  # Changed to 3 channels
model.compile(
    optimizer=Adam(learning_rate=LEARNING_RATE),
    loss=BinaryCrossentropy(label_smoothing=SMOOTHING),  # Use built-in function
    metrics=['accuracy', tf.keras.metrics.AUC(name='auc')]
)

# Print model summary
print(model.summary())

# Callbacks
class ClearMemory(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs=None):
        gc.collect()
        tf.keras.backend.clear_session()

clear_memory = ClearMemory()

early_stopping = EarlyStopping(monitor='val_auc', patience=10, restore_best_weights=True, mode='max')
reduce_lr = ReduceLROnPlateau(monitor='val_auc', factor=0.5, patience=5, min_lr=1e-7, verbose=1, mode='max')
model_checkpoint = ModelCheckpoint('best_hybrid_model.keras', monitor='val_auc', mode='max', save_best_only=True, verbose=1)

# Train Model
history = model.fit(
    train_gen,
    validation_data=test_gen,
    epochs=EPOCHS,
    callbacks=[early_stopping, reduce_lr, model_checkpoint, clear_memory]
)

# Save the model
model.save('hybrid_model.keras')
print("Model saved as hybrid_model.keras")

# Evaluate Model
test_preds = model.predict(test_gen)
test_preds = (test_preds > 0.5).astype(int)  # Convert probabilities to binary predictions
```
